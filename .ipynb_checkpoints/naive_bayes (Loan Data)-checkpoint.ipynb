```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('loan_data.csv')
df.head()
```

```python
pre_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = pre_df.drop('not.fully.paid', axis=1)
y = pre_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train);

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)

y_pred = model.predict(X_test)

accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import numpy as np
import random

# Load the loan data
loan_data = pd.read_csv('loan_data.csv')

# Create a label encoder for the purpose column
le = LabelEncoder()
loan_data['purpose'] = le.fit_transform(loan_data['purpose'])
print(loan_data['purpose'].head(8))

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(loan_data, loan_data['not.fully.paid'], test_size=0.2)

# Create a naive Bayes classifier
clf = GaussianNB()

# Train the classifier on the training set
clf.fit(X_train, y_train)

# Predict the loan status for the test set
y_pred = clf.predict(X_test)

score=clf.score(X_test, y_test)

# Evaluate the performance of the classifier
print('accuracy of the naive bayes classifier is: ', score)
```


A naive Bayes classifier is a simple probabilistic classifier based on Bayes' theorem. It is a supervised learning algorithm 
that is used for classification tasks. The naive Bayes classifier assumes that the features are independent of each other,
which means that the probability of a particular class label can be calculated by multiplying the probabilities of the 
individual features.

The naive Bayes classifier is a generative classifier, which means that it models the probability distribution of the features. This makes it a good choice for classification tasks where the features are not independent of each other.

The naive Bayes classifier is a fast and easy to train classifier. It is also a scalable classifier, which means that it can be used to classify large datasets.

However, the naive Bayes classifier is not a robust classifier. It is sensitive to the presence of noise in the data.

```python
pre_df= pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
from sklearn.model_selection import train_test_split

x=pre_df.drop('not.fully.paid',axis=1)
y=pre_df['not.fully.paid']

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

model= GaussianNB()

model.fit(x_train,y_train)

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report
)

y_pred= model.predict(x_test)

accuracy= accuracy_score(y_pred,y_test)
f1= f1_score(y_pred,y_test,average='weighted')

print('accuracy: ', accuracy)
print('f1score: ', f1)
```
