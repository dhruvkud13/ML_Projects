```python
import pandas as pd
df= pd.read_excel('World Bank Indicators.xlsx')
df.head(15)
```

```python
# Select the rows where the gender is female
female_df = df[df['Health: Life expectancy at birth, female (years)'].notnull()]

# Calculate the average life expectancy at birth for women
average_life_expectancy = female_df['Health: Life expectancy at birth, female (years)'].mean()

# Print the average life expectancy at birth for women
print(average_life_expectancy)
```

```python
# Calculate the correlation between the number of mobile phone subscribers and the number of internet users
correlation = df['Business: Mobile phone subscribers'].corr(df['Business: Internet users (per 100 people)'])

# Print the correlation coefficient
print(correlation)
```

```python
# Calculate the average GDP per capita for urban areas
urban_gdp_per_capita = df[df['Population: Urban (count)'].notnull() > 0]['Finance: GDP per capita (current US$)'].mean()

# Calculate the average GDP per capita for rural areas
rural_gdp_per_capita = df[df['Population: Urban (count)'].notnull() == 0]['Finance: GDP per capita (current US$)'].mean()

# Calculate the difference in the average GDP per capita
difference_in_gdp_per_capita = urban_gdp_per_capita - rural_gdp_per_capita

# Print the difference in the average GDP per capita
print(difference_in_gdp_per_capita)
```

```python
# Sort the DataFrame by the number of passenger cars per 1,000 people
df_new= df.dropna(subset=['Transit: Passenger cars (per 1,000 people)'])
df_new.head()
df_new = df_new.sort_values('Transit: Passenger cars (per 1,000 people)', ascending=False)

# Get the top five countries
top_five_countries = df_new.head(5)

# Print the top five countries
print(top_five_countries)
```

```python
import matplotlib.pyplot as plt
# Calculate the percentage of the population that is 0-14 years old
percentage_0_14 = df['Population: Ages 0-14 (% of total)'].mean()

# Calculate the percentage of the population that is 15-64 years old
percentage_15_64 = df['Population: Ages 15-64 (% of total)'].mean()

# Calculate the percentage of the population that is 65+ years old
percentage_65_plus = df['Population: Ages 65+ (% of total)'].mean()

# Calculate the GDP per capita
gdp_per_capita = df['Finance: GDP per capita (current US$)'].mean()

# Create a scatter plot of the percentage of the population that is 0-14 years old vs. the GDP per capita
plt.scatter(percentage_0_14, gdp_per_capita)
plt.scatter(percentage_15_64, gdp_per_capita)
plt.scatter(percentage_65_plus, gdp_per_capita)
plt.xlabel('Percentage of population that is 0-14 years old')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# Sort the DataFrame by the life expectancy at birth
df = df.sort_values('Health: Life expectancy at birth, female (years)', ascending=False)

# Get the top 5 countries
top_five_countries = df.head(5)

# Print the top five countries
print(top_five_countries)
```

```python
# Sort the DataFrame by the GDP per capita
df = df.sort_values('Finance: GDP per capita (current US$)', ascending=False)

# Get the top 1 country
highest_gdp_per_capita = df.head(1)

# Get the bottom 1 country
lowest_gdp_per_capita = df.tail(1)

# Print the top 1 country
print(highest_gdp_per_capita)

# Print the bottom 1 country
print(lowest_gdp_per_capita)
```

```python
# Calculate the mean GDP per capita
mean_gdp_per_capita = df['Finance: GDP per capita (current US$)'].mean()

# Calculate the standard deviation of GDP per capita
standard_deviation_gdp_per_capita = df['Finance: GDP per capita (current US$)'].std()

# Print the mean and standard deviation of GDP per capita
print('Mean GDP per capita:', mean_gdp_per_capita)
print('Standard deviation of GDP per capita:', standard_deviation_gdp_per_capita)

# Create a box plot of the GDP per capita
df.boxplot('Finance: GDP per capita (current US$)')
plt.show()
```
