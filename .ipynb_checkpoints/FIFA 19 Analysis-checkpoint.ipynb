```python
import pandas as pd
```

```python
df= pd.read_csv("FIFA19.CSV")
df.head()
```

```python
country_counts={}
for country in df["Nationality"]:
    if country not in country_counts:
        country_counts[country]=0
    else:
        country_counts[country]+=1
        
most_players= sorted(country_counts.items(), key= lambda x: x[1], reverse=True)[:5]

for country, count in most_players:
    print(f"{country}:{count}")
```

```python
crossing_players= df[df["Crossing"] > 80]
number_of_players= crossing_players.shape[0]

print("there are {} players with crossing above 80".format(number_of_players))
```

```python
average_heading= df["HeadingAccuracy"].mean()

print("the average heading accuracy is: ", average_heading)
```

```python
df1= df.query("Overall>=88")
df1.head()

df1['Value']= df1['Value'].str.replace('€','')
df1['Value']= df1['Value'].str.replace('M','')
df1['Wage']=df1['Wage'].str.replace('€','')
df1['Wage']=df1['Wage'].str.replace('K','')

df1.head()

df1.Value= df1.Value.astype('float')
df1.Wage= df1.Wage.astype('int')
df1.Name= df1.Name.astype('category')

df1= df1.sort_values(['Value','Wage'], ascending=False)

for index, row in df1.iloc[:5].iterrows():
    print(f"{row['Name']}, {row['Value']}, {row['Wage']}")
```
