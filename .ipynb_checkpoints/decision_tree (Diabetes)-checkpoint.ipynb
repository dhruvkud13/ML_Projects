```python
import pandas as pd
data= pd.read_csv('diabetes.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(data.loc[:, data.columns!= 'Outcome'], data['Outcome'],test_size=0.2)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
gini= DecisionTreeClassifier(criterion="gini")
gini.fit(x_train, y_train)
y_pred_gini= gini.predict(x_test)
accuracy_gini= accuracy_score(y_test, y_pred_gini)
```

```python
entropy= DecisionTreeClassifier(criterion="entropy")
entropy.fit(x_train,y_train)
y_pred_entropy= entropy.predict(x_test)
accuracy_entropy= accuracy_score(y_test, y_pred_entropy)
```

```python
depth3= DecisionTreeClassifier(criterion='gini',max_depth=3)
depth3.fit(x_train,y_train)
y_pred_depth3= depth3.predict(x_test)
accuracy_depth3= accuracy_score(y_test, y_pred_depth3)
```

```python
knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred_knn= knn.predict(x_test)
accuracy_knn= accuracy_score(y_test, y_pred_knn)
```

```python
print('accuracy of decision tree with gini index is: ', accuracy_gini)
print('accuracy of decision tree with entropy is: ',accuracy_entropy)
print('accuracy of decision tree with depth 3 is: ',accuracy_depth3)
print('accuracy of the knn classifier is: ', accuracy_knn)
```

```python
import numpy as np
test_data= np.array([5,150,70,33,0,33,0.63,51])
prediction1= gini.predict(test_data.reshape(1,-1))
prediction2= entropy.predict(test_data.reshape(1,-1))
prediction3= depth3.predict(test_data.reshape(1,-1))
prediction4= knn.predict(test_data.reshape(1,-1))

print('prediction using gini is: ',prediction1)
print('prediction using entropy is: ',prediction2)
print('prediction using depth3 is: ',prediction3)
print('prediction using knn is: ',prediction4)
```
