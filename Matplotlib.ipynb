```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,90,100)
y= x**2 + 2*x
z = np.sin(y)

plt.plot(x,y, label="first func")
plt.plot(x,z, label="second func")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

marks= np.random.randint(0,20,60)
# marks=[1,2,3,4,4,5,5,6,7,20]
print(marks)

plt.hist(marks, bins=20)
plt.xlabel("marks")
plt.ylabel("no of students")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(-2*np.pi,2*np.pi,100)
plt.subplot(221)
plt.plot(x,np.sin(x), label="sinx")
plt.xlabel("x")
plt.ylabel("sinx")

plt.subplot(222)
plt.plot(x,np.cos(x), label="cosx")
plt.xlabel("x")
plt.ylabel("cosx")

plt.subplot(223)
plt.plot(x,np.tan(x), label="tanx")
plt.xlabel("x")
plt.ylabel("tanx")

plt.subplot(224)
plt.plot(x,np.tan(x), label="tanx")
plt.xlabel("x")
plt.ylabel("tanx")

plt.legend()

plt.show()
```
