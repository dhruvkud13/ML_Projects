```python
def most(x):
    counts={}
    for i in range(len(x)):
#         print(x[i])
        if x[i] in counts:
            counts[x[i]]+=1
        else:
            counts[x[i]]=1
    max_count= max(counts.values())
    most_frequent_items=[]
    for key, value in counts.items():
        if value== max_count:
          most_frequent_items.append(key)
    return most_frequent_items, max_count
most_freq= most(zomato['cuisines'])
print(most_freq)           
```

```python
import pandas as pd
import nltk

# Create a Pandas DataFrame from the CSV data
# df = pd.read_csv('restaurants.csv')

zomato['dish_liked'] = zomato['dish_liked'].astype(str)

# Split the `dish_liked` column on spaces
dish_words = zomato['dish_liked'].str.split(',', expand=True)

# Count the number of times each word appears in the `dish_liked` column, ignoring non-numeric columns
word_counts = dish_words.sum(numeric_only=True)

# Identify the most common words
most_common_words = word_counts.sort_values(ascending=False).head(10)

# Look at the dishes that contain these words
for word in most_common_words.index:
    print(word, zomato[zomato['dish_liked'].str.contains(word)]['name'].to_list())
```
